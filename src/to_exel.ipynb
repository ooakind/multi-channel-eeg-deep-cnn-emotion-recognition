```python
import os
import pandas as pd
import pickle as pkl
import numpy as np

# For 10 fold CV
# col = ["Task", "Metrics", "Rhythm", "Fold-1", "Fold-2", "Fold-3", "Fold-4", "Fold-5", "Fold-6", "Fold-7", "Fold-8", "Fold-9", "Fold-10", "Average", "Std"]

# For 8 fold CV
col = ["Task", "Metrics", "Rhythm", "Fold-1", "Fold-2", "Fold-3", "Fold-4", "Fold-5", "Fold-6", "Fold-7", "Fold-8", "Average", "Std"]

# save_file = './results_k-foldCV_unseen_trials.csv'
save_file = './results_10-foldCV_unseen_subjects.csv'
# read_path = '/DEAP_k-foldCV_unseen_trials_results'#
read_path = '/DEAP_10-foldCV_unseen_subjects_results'

try:
    df = pd.read_csv(save_file)
except:
    df = pd.DataFrame(columns=col)

emotions = ['arousal', 'dominance', 'valence']
bands = ['delta', 'theta', 'alpha', 'beta', 'gamma', 'all']

for e in emotions:
    for b in bands:
        filename = f'./eeg_data{read_path}/DEAP_{e}_{b}.pkl'
        if os.path.exists(filename):
            with open(filename, 'rb') as f:
                res = pkl.load(f)
            acc = np.array(res['accuracy'])[:,1]
            values = np.concatenate(([e, 'Accuracy', b], acc, [np.average(acc), np.std(acc)]), axis=0)
            new_row = dict(zip(col, values))
            df = df.append(new_row, ignore_index=True)

df.to_csv(save_file, index=False)
```
